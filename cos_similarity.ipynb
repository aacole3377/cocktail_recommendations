```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('onehot.csv')
orig_df = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-05-26/boston_cocktails.csv")
df
```

```python
df.reset_index(inplace=True)
df.set_index('name', inplace=True)
df.index
```

```python
similarity_matrix = cosine_similarity(df)
similarity_matrix
```

```python
def get_similar_drinks(drink_name, n=5):
    if drink_name not in df.index:
        return f"Drink '{drink_name}' not found in the dataset."
    
    # Get the index of the input drink
    drink_index = df.index.get_loc(drink_name)
    
    # Get the similarity scores for the input drink
    similarity_scores = similarity_matrix[drink_index]
    
    # Get the indices of the top n most similar drinks (excluding the input drink itself)
    similar_indices = similarity_scores.argsort()[-n-1:-1][::-1]
    
    # Get the names of the similar drinks
    similar_drinks = df.index[similar_indices]
    
    return similar_drinks
```

```python
similar_drinks = get_similar_drinks('Martini')
similar_drinks
```
